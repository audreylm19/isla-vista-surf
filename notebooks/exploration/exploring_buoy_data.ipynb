````python
timelist = ['YEAR','MO','DY','HR']
s19[timelist] = s19[timelist].astype(str)
````

Need to make the string the correct format, ```YYYYMMDDHHMM```

````python
#month formatting

for i in range(len(s19)):
    if len(s19['MO'][i])<2:
        s19['MO'][i]='0'+s19['MO'][i]

#day formatting

for i in range(len(s19)):
    if len(s19['DY'][i])<2:
        s19['DY'][i]='0'+s19['DY'][i]

#hour formatting

for i in range(len(s19)):
    if len(s19['HR'][i])<2:
        s19['HR'][i]='0'+s19['HR'][i]

s19['UTC'] = s19['YEAR']+s19['MO']+s19['DY']+s19['HR']+'00'

s19['UTC'] = pd.to_datetime(s19['UTC'], utc=True)

s19 = s19.drop(columns=timelist)
s19.head()
````
